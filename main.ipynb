```python
import matplotlib.pyplot as plt
import numpy as np
import utils
```

Вычисляем истинные вероятности для каждого радиуса

```python
a = 1
radiuses = utils.get_uniform_radius(a)
get_true_probability_vec = np.vectorize(utils.get_true_probability)
true_probabilities = get_true_probability_vec(a, radiuses)
true_probabilities
```

```python
n = 300 # Кол-во точек
number_of_points = np.arange(n) + 10
```

Дальше идут пары графиков экспериментально выведенных вероятностей и отклонений от истинных значений

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
count_gen = np.array([utils.count_generated_points(a, radiuses[0], i)/i for i in number_of_points])

ax[0].plot(number_of_points, count_gen, label='Экспериментальное значение', color='purple')
ax[0].plot(number_of_points, np.array([true_probabilities[0] for _ in range(len(number_of_points))]), label='Истинное значение', color='orange')

ax[0].set_xlabel('Кол-во точек')
ax[0].set_ylabel('Доля точек попавших в круг')
ax[0].set_title(f"Доли точек попавших в круг радиусом {radiuses[0]} в зависимости от количества точек", fontsize=8)
ax[0].legend(fontsize=8)
ax[0].grid()

ax[1].plot(number_of_points, abs(count_gen - true_probabilities[0]), label='Ошибка', color='red')

ax[1].set_xlabel('Кол-во точек')
ax[1].set_ylabel('Значение ошибки')
ax[1].set_title(f"Значение ошибки для круга радиусом {radiuses[0]} в зависимости от количества точек", fontsize=8)
ax[1].legend(fontsize=8)
ax[1].grid()

plt.show()
```
